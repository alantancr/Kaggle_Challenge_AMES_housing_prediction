# Check the correlation between the features and the target variable

# Data preparation

```python
#merge the train and test data and apply the same processing method to be consistent
df_train['dataset_type']='train'
df_test['dataset_type']='test'

df_combined=pd.concat((df_train,df_test),sort=False)
```

```python
df_combined.reset_index(drop=True)
```

```python
print("Shape of combined dataframe: {}".format(df_combined.shape))
```

```python
df_combined.drop(['pid'],axis=1,inplace=True)
```

```python
# according to data document, remove houses sqft>4000
mask_area= df_combined['gr_liv_area']>4000
df_combined[mask_area]
```

```python
df_combined=df_combined[~mask_area]
```

```python
#there were 5 rows removed
print("Shape of combined dataframe: {}".format(df_combined.shape))
```

```python
#fill nan/empty cells with na

cols=['pool_qc','bsmt_qual','fireplace_qu','garage_qual','garage_cond','bsmtfin_type_1','bsmtfin_type_2','alley'
      ,'mas_vnr_type','bsmt_cond','bsmt_exposure','garage_type','garage_finish','fence','misc_feature']
for col in cols:
    df_combined[col]=df_combined[col].fillna(value='na')
```

```python
#fill nan/empty cell with 0

cols =['bsmt_half_bath','bsmt_full_bath','mas_vnr_area','bsmtfin_sf_1','bsmtfin_sf_2','bsmt_unf_sf'
       ,'total_bsmt_sf','garage_yr_blt','garage_cars','garage_area']
for col in cols:
    df_combined[col]=df_combined[col].fillna(value='0')
```

```python
# filling col with median value
df_combined["lot_frontage"] = df_combined.groupby("neighborhood")["lot_frontage"].transform(lambda x: x.fillna(x.median()))
```

```python
#check if there are still null values and handle them accordingly
nan_colnames_new = pd.DataFrame(df_combined.isnull().sum(),columns = ['count'])
nan_colnames_new = nan_colnames_new[nan_colnames_new['count'] > 0]
nan_colnames_new
```

```python
def outliers(df):
    outlier_cols = []
    for column in df.columns.to_list():
        if df[column].dtype in ('int64', 'float64'):
            if df[column].max() > df[column].mean() + 3*df[column].std(ddof=0) \
            or df[column].min() < df[column].mean() - 3*df[column].std(ddof=0):
                outlier_cols.append(column)
    return outlier_cols
```

```python
columns_list = outliers(df_combined)
```

```python
columns_list.remove('saleprice')
```

```python
#
def subplot_scatterplots(dataframe, list_of_columns, target_var):
    nrows = int(np.ceil(len(list_of_columns)/3)) # Plot in 3 columns
    fig, ax = plt.subplots(nrows=nrows, ncols=3,figsize=(15,50)) 
    plt.subplots_adjust(bottom=0.1, top=0.9)
    ax = ax.ravel() 
    for i, column in enumerate(list_of_columns): 
        sns.scatterplot(x=dataframe[column], y=target_var, data=dataframe, ax=ax[i])
```

```python
subplot_scatterplots(df_combined,columns_list,'saleprice')
```
